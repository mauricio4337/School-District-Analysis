```python
# Add Pandas and os packages
import pandas as pd
import os

# Save paths to load data files
student_data_load = os.path.join(".","Resources","students_complete.csv")

# Read the student data file and store it in a Pandas DataFrame.
student_data_df = pd.read_csv(student_data_load)

# Put the student names in a list.
student_names = student_data_df["student_name"].tolist()
# student_names

# Create a new list and use it for the for loop to iterate through the list.
students_to_fix = []

for name in student_names:
    if len(name.split()) >= 3:
        students_to_fix.append(name)

# Get the number of students whose names are greater than or equal to 3
len(students_to_fix)

# Add prefixes less than or equal to 4 to prefixes
# Add the suffixes less than or equal to 3 to a new list.

prefixes = []
suffixes = []
for name in students_to_fix:
    if len(name.split()[0]) <= 4:
        prefixes.append(name.split()[0])
    if len(name.split()[-1]) <= 3:
        suffixes.append(name.split()[-1])

# Get the unique items in prefixes
# print(set(suffixes))

# Declare a list of prefixes and suffixes
prefixes_suffixes = ['Dr. ', 'Miss ', 'Mr. ', 'Mrs. ', 'Ms. ', 
                     ' DVM', ' PhD', ' MD', ' DDS']

for part_of_name in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(part_of_name, "")

# Put cleaned student names in a list.
cleaned_names = student_data_df["student_name"].tolist()
len(cleaned_names)

# Create a new list and use it for the for loop to iterate through the list.
students_fixed = []

# Use an if statement to check the length of the name.

# If the name is greater than or equal to 3, add the name to the list.

for name in cleaned_names:
    if len(name.split()) >= 3:
        students_fixed.append(name)

# Get the length of the students' names that are greater than or equal to 3.
students_fixed
    
# print(len(suffixes))
# len(prefixes)

    
```
